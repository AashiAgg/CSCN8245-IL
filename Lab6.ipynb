### Question 1- DATA INTEGRATION:  (5)
Check the files inside microbiome folder, (MID1, MID2, ... , MID9.xls files), you need to combine those files into single dataframe.<br>

```python
import pandas as pd
import os
from scipy.stats import chi2_contingency
import glob
```

```python
# Find all Excel files in the directory
path = 'microbiome'
files = glob.glob(os.path.join(path, 'MID*.xls'))

# Read each Excel file into a dataframe and add a column of barcodes
data = []
for file in files:
    df = pd.read_excel(file, header=None)
    barcode = os.path.splitext(os.path.basename(file))[0]
    df['BARCODE'] = barcode
    data.append(df)

# Combine the dataframes into a single dataframe
all_data_df = pd.concat(data, ignore_index=True)

# Read the metadata file
metadata = pd.read_csv(os.path.join(path, 'metadata.csv'))

# Merge the combined data with the metadata on the 'BARCODE' column
finalData = all_data_df.merge(metadata, on='BARCODE', how='left')

# Display the result
finalData.head()
```
